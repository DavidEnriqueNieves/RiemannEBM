```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

out = widgets.Output()

def plot(a=1.0, freq=1.0):
    with out:
        out.clear_output(wait=True)

        x = np.linspace(0, 10, 400)
        y = a * np.sin(freq * x)

        plt.figure(figsize=(5, 3))
        plt.plot(x, y)
        plt.ylim(-5, 5)
        plt.show()

ui = widgets.interact(
    plot,
    a=widgets.FloatSlider(1.0, min=0, max=5, step=0.1),
    freq=widgets.FloatSlider(1.0, min=0.1, max=5, step=0.1)
)

display(out)
```
